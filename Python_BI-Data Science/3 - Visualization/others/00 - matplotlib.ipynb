# Basic Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
#Object Oriented Method
fig = plt.figure()

# left, bottom, width, height 
axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X_Label')
axes.set_ylabel('Y_Label')
axes.set_title('My Title')
```

```python
fig = plt.figure()

# left, bottom, width, height
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.5,0.2,0.4,0.3])

#fig = plt.figure()

#axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
#axes2 = fig.add_axes([0.5,0.2,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Large')

axes2.plot(y,x)
axes2.set_title('Small')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')
```

# Figure Size and DPI (Dots per inch)

```python
fig = plt.figure(figsize=(3,2), dpi = 100)

ax = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure(figsize=(8,2), dpi = 100)

ax = fig.add_axes([0,0,1,1])
```

```python
fig,axes = plt.subplots(figsize=(8,2))

axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=1,figsize=(8,2))

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('mychart.png', dpi=200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_title('Title')
ax.set_ylabel('Y Label')
ax.set_xlabel('X Label')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Square')
ax.plot(x,x**3, label = 'X Cube')

ax.legend(loc=0) # 1 to 10, check out the matplotlib page
#ax.legend(loc=(0.1,0.1))
```

# Plot Apperance

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', linewidth=3, alpha = 0.5) #alpha is transperency
#ax.plot(x,y,color='green', lw=3, alpha = 0.5) #even lw works 
#ax.plot(x,y,color='#FFBC00') #RGB Hex Code
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', lw=3, linestyle = '--') 
#ax.plot(x,y,color='green', lw=3, linestyle = '-.') # ls also works for linestyle
#ax.plot(x,y,color='green', lw=3, linestyle = 'steps') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,y,color='green', lw=3, ls = '-', marker = 'o') 
#ax.plot(x,y,color='green', lw=3, ls = '-', marker = '*') 
#ax.plot(x,y,color='green', lw=3, ls = '-', marker = 'o', markersize = 5) 
ax.plot(x,y,color='green', lw=3, ls = '-', marker = 'o', markersize = 10, markerfacecolor = 'yellow', 
markeredgewidth = 3, markeredgecolor = 'green') 
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='green', lw=2, ls = '--') 

ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
