```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6]

y = [1,2,3,4,5,6]

plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Our first Graph")


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0.0,2.0,0.01)

s = 1 + np.cos(2*np.pi*t)

plt.plot(t,s,'--')
plt.grid()
plt.xlabel("Time(t)")
plt.ylabel("Voltage(V)")
plt.title("Cosine wave plot(Cos(x))")

plt.show()
```
